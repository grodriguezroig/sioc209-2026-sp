```python
import numpy as np

# --- Test Keras 3 (multi-backend) ---
import keras
print(f"Keras {keras.__version__}  (backend: {keras.backend.backend()})")

cifar = keras.datasets.cifar100
(x_train, y_train), (x_test, y_test) = cifar.load_data()
x_train, x_test = x_train.astype("float32") / 255.0, x_test.astype("float32") / 255.0

model = keras.Sequential([
    keras.layers.Input(shape=(32, 32, 3)),
    keras.layers.Conv2D(32, 3, activation="relu"),
    keras.layers.GlobalAveragePooling2D(),
    keras.layers.Dense(100),
])
model.compile(optimizer="adam",
              loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
model.fit(x_train[:2000], y_train[:2000], epochs=2, batch_size=64, verbose=1)
print("✓ Keras model trained successfully\n")

# --- Test PyTorch ---
import torch
print(f"PyTorch {torch.__version__}  (MPS available: {torch.backends.mps.is_available()})")
t = torch.randn(2, 3)
print(f"✓ PyTorch tensor: {t.shape}\n")

# --- Test scientific stack ---
import xarray, cartopy, matplotlib, sklearn, pandas, scipy
print(f"✓ xarray {xarray.__version__}, cartopy {cartopy.__version__}, "
      f"sklearn {sklearn.__version__}, pandas {pandas.__version__}")
print("\nAll packages OK — environment is ready!")
```
